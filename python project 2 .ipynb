```python
df=df.drop(columns=["Unnamed: 0"])
df
```

```python
df["Date"]=pd.to_datetime(df["Date"])
df
```

```python
df["weekly_ma"]=df['Sales'].rolling(window=7).mean()
df["monthly _ma"]=df['Sales'].rolling(window=30).mean()
df
```

```python
df.head(15)
```

```python
weekly_mae = (df['Sales'] - df['weekly_ma']).abs().mean()
monthly_mae = (df['Sales'] - df['monthly _ma']).abs().mean()
df
```

```python
print(f"weekly_mae: {weekly_mae:.2f}")
print(f"monthly_mae: {monthly_mae:.2f}")
```

```python
df["Date"]=pd.to_datetime(df["Date"])
df
```

```python
plt.figure (figsize=(12,6))
plt.plot(df.index,df["Sales"],label="actual sales",color="blue",alpha=0.6)
plt.plot(df.index,df["weekly_ma"],label=" 1 week moving average",color="orange",linewidth=2)
plt.plot (df.index,df["monthly _ma"],label="1 month moving average",color="green",linewidth=2)

plt.title("slaes forecasting")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid(True)
plt.tight_layout()
plt.show()
```
