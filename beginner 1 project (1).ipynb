```python
for col in data_set.columns:
    data_set[col]=pd.to_numeric(data_set[col],errors="ignore")
data_set[col]
```

```python
price=["discounted_price"]
for col in price:
    data_set[col]=pd.to_numeric(data_set["discounted_price"],errors="ignore")
data_set[col]
```

```python
data_set[col].astype(str).str.replace(r"[₹,]", "",regex=True).str.strip()
```

```python
price_cols = ["discounted_price"]

for col in price_cols:
    data_set[col] = (
        data_set[col]
        .astype(str)
        .str.replace(r"[₹,]", "", regex=True)
        .str.strip()
    )
data_set[col]=pd.to_numeric(data_set["discounted_price"],errors="ignore")
```

```python
data_set[col].dtype
```

```python
discounted_percentage = ["discount_percentage"]

for col in discounted_percentage :
    data_set["discount_percentage"] = (
        data_set["discount_percentage"]
        .astype(str)
        .str.replace(r"[%,]", "", regex=True)
        .str.strip()
        
    )
data_set["discount_percentage"]=(pd.to_numeric(data_set["discount_percentage"],errors="ignore")/100)
```

```python
data_set["discount_percentage"]
```

```python

data_set["rating_count"] = (
    data_set["rating_count"]
    .astype(str)
    .str.replace(",", "", regex=False)
)

data_set["rating_count"] = pd.to_numeric(
    data_set["rating_count"], errors="coerce"
)
```

```python
data_set["rating_count"]
```

```python
data_set["rating"] = (
    data_set["rating"]
    .astype(str)
    .str.replace(",", "", regex=False)
)

data_set["rating"] = pd.to_numeric(
    data_set["rating"], errors="coerce"
)
```

```python
data_set["rating"]
```

```python
text_cols= ['product_id', 'product_name', 'category',
       'actual_price',
       'about_product', 'user_id', 'user_name', 'review_id', 'review_title',
       'review_content', 'img_link', 'product_link']
```

```python
data_set["actual_price"].dtype
```

```python
actual_price= ["actual_price"]

for act in actual_price:
    data_set[act] = (
        data_set[act]
        .astype(str)
        .str.replace(r"[₹,]", "", regex=True)
        .str.strip()
    )
data_set[act]=pd.to_numeric(data_set["actual_price"],errors="ignore")
```

```python
import matplotlib.pyplot as plt

data_sorted = data_set.sort_values(by="rating", ascending=False)

plt.figure(figsize=(100, 5))
plt.bar(data_sorted["product_name"], data_sorted["rating"])
plt.xticks(rotation=90)
plt.title("Product Ratings (Sorted)")
plt.tight_layout()
plt.show()
```

```python
top10 = data_set.sort_values(by="rating", ascending=False).head(10)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.barh(top10["product_name"], top10["rating"])

plt.xlabel("Rating")
plt.ylabel("Product")
plt.title("Top 10 Products by Rating")

plt.gca().invert_yaxis()  # highest rating on top
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.plot(top10["product_name"], top10["rating"], marker='o', linestyle='-', color='b')

plt.xlabel("Product")
plt.ylabel("Rating")
plt.title("Top 10 Products by Rating")
plt.xticks(rotation=45, ha="right")  # rotate labels for readability

plt.grid(True)  # optional, makes chart easier to read
plt.tight_layout()
plt.show()
```
